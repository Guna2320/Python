## Feature Engineering

```python
# Checking the ouliers
for col in df.select_dtypes('number').columns:
    print(col)
    print(df[col].describe([0.01, 0.05, 0.25, 0.5, 0.75, 0.95, 0.99]))
```

```python
# Removing outliers
for col in df.select_dtypes('number').columns: 
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3-q1 #Interquartile range 
    fence_low  = q1-1.5*iqr 
    fence_high = q3+1.5*iqr
    df.loc[df[col] < fence_low, col] = fence_low
    df.loc[df[col] > fence_high, col] = fence_high
```

## SMOTE
### implementing SMOT framework to make the data balanced

```python
!pip install -U scikit-learn==1.0.2 --prefer-binary
!pip install -U imbalanced-learn
import sklearn
import imblearn
print("scikit-learn version:", sklearn.__version__)
print("imbalanced-learn version:", imblearn.__version__)
```

```python

from imblearn.over_sampling import SMOTE

# Assuming df is your DataFrame with features and target variable
X = df.drop("y", axis=1)  # Features
y = df["y"]  # Target variable

# Instantiate SMOTE
smote = SMOTE(random_state=42)

# Apply SMOTE to generate synthetic samples
X_resampled, y_resampled = smote.fit_resample(X, y)

# Create a DataFrame with the resampled data
df_resampled = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.Series(y_resampled, name="y")], axis=1)

```

```python
df_resampled.head(4)
```

##  Checking multicollinearity using VIF:

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
X = df_resampled.drop('y', axis=1)  
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)  
vif = vif.sort_values(by = "VIF", ascending = False)
print(vif)
```

```python
# pdays columns to be removed because it has high VIF score it may affect the model prediction 
columns_to_remove = ['pdays']
df_resampled = df_resampled.drop(columns=columns_to_remove)
```

## Recursive Feature Elimination with Cross-Validated selection
RFECV  is an extension of RFE that automatically performs cross-validated feature selection and estimates the optimal number of features 

```python
from sklearn.feature_selection import RFECV
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold

# Assuming X is your feature matrix and y is your target variable
X = df_resampled.drop('y', axis=1)
y = df_resampled['y']

# Replace RandomForestClassifier with the model of your choice
model = RandomForestClassifier()

# Use stratified k-fold cross-validation for classification problems
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)  # Adjust the number of splits as needed

# Initialize RFECV
rfecv = RFECV(estimator=model, step=1, cv=cv, scoring='accuracy', n_jobs=-1)
rfecv.fit(X, y)

# Selected features
selected_features = X.columns[rfecv.support_]

# Print the results
print("Optimal Number of Features:", rfecv.n_features_)
print("Selected Features:", selected_features)
```

```python
# Assuming df_resampled is your original DataFrame
selected_features_df = df_resampled[selected_features.tolist() + ['y']]

# Display the head of the selected features DataFrame
print(selected_features_df.head(4))
```

# Random forest for Resampled data 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

# Assuming df_resampled is your original DataFrame
X = df_resampled.drop('y', axis=1)
y = df_resampled['y']

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

# Define the Random Forest classifier
rf_classifier = RandomForestClassifier(random_state=42)
```

```python
X.head(3)
```

### Hyper parameter tunning

```python
# Initialize RandomizedSearchCV with reduced iterations
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

# Define the parameter grid for RandomizedSearchCV
param_dist = {
    'n_estimators': randint(10, 200),
    'max_features': ['auto', 'sqrt', 'log2', None],
    'max_depth': randint(1, 20),
    'min_samples_split': randint(2, 20),
    'min_samples_leaf': randint(1, 20),
    'bootstrap': [True, False]
}

# Initialize RandomizedSearchCV with reduced iterations
random_search = RandomizedSearchCV(
    rf_classifier, param_distributions=param_dist, n_iter=20, cv=5, random_state=42, n_jobs=-1
)
```

```python
#Fit RandomizedSearchCV to the data
random_search.fit(X_train, y_train)
```

```python
# Print the best parameters found
print("Best Parameters:", random_search.best_params_)
```

```python
# Create a RandomForestClassifier with the best hyperparameters
best_rf_classifier = RandomForestClassifier(
    bootstrap=False,
    max_depth=16,
    max_features='log2',
    min_samples_leaf=5,
    min_samples_split=4,
    n_estimators=181,
    random_state=42
)
```

```python
# Fit the classifier on the training data
best_rf_classifier.fit(X_train, y_train)
```

```python
# Making predictions on the test set
y_pred = best_rf_classifier.predict(X_test)
```

```python
# Evaluate the performance
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)
```

```python
print("Accuracy:", accuracy)
```

```python
print("\nConfusion Matrix:\n", conf_matrix)
```

```python
print("\nClassification Report:\n", classification_rep)
```

```python
# ROC Curve and AUC for the best model
y_proba_rf = best_rf_classifier.predict_proba(X_test)[:, 1]
fpr_rf, tpr_rf, thresholds_rf = roc_curve(y_test, y_proba_rf)
roc_auc_rf = auc(fpr_rf, tpr_rf)

# Plot ROC curve for the best model
plt.figure(figsize=(8, 6))
plt.plot(fpr_rf, tpr_rf, color='darkorange', lw=2, label='ROC curve (AUC = {:.2f})'.format(roc_auc_rf))
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve - Best Model')
plt.legend(loc='lower right')
plt.show()
```
